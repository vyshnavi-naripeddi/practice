 Verify year is a leap or not
- it should be divisible by 4
- it should not divisible by 100
- divisible by 400

```python
year = int(input("enter a year: "))
if(year%400==0) or (year%4==0 and year%100!=0):
    print(year,"is a leapyear")
else:
    print(year,"is not a leap year")
    
```

# while loop


```python
i =1
while i<101:
    print(i,end='\t')
    i +=1
```

- break:to terminate the loop
- continue:skip the current execution of the loop



```python
print the numbers from 1 to 100 whenever 50 stop
```

```python
for travel in range(1,101,1):
    if travel == 50:
        break
    print(travel)
```

```python
for travel in range(1,101,1):
    if travel == 50:
        break
    elif travel%2==0:
        continue
    print(travel)
```

```python


while True:
    inp = input()
    if inp == 'cancel':
        print('loop terminated')
        break
    print(inp)
```
